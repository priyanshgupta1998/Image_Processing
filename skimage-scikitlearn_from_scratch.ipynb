#2. Changing Image Format

```
from skimage.color import rgb2gray
img = imread('tiger.jpeg')  # original colored image
img_new = rgb2gray(img)   # convert  colored image into gray image

plt.subplot(121)
imshow(img)
plt.title('RGB Format') 

plt.subplot(122)
imshow(img_new)
plt.title('Grayscale Format') 

plt.show()
```

`Other two popular formats are HSV (hue, saturation, value) and HSL (hue, saturation, lightness) which are alternative representations of the RGB format. `

`1. Hue is a degree on the color wheel where 0 is for red, 120 is green, 240 is blue and again 360 would be red.`     
`2. Saturation represents the percentage of that color, where 0 is white and 100  is the full color.`    
`3. Value denotes the mixture of the colors with varying amounts of black or white paint.`     
`4. Lightness is another way to show the shade of the image where 0 is black and 1 is white.(means either color is fade or shine)`

#3. Image resizing
`Ideally, the size of the images should be the same when were building our model. If were using a pre-trained model, it is important to resize and normalize the input data to the same format over which the network was originally trained. This is why resizing images is an important image preprocessing step.`

```
from skimage.transform import resize
img = imread('tiger.jpeg')
#resize image
img_resized = resize(img, (300, 300))

#plot images
plt.subplot(121), imshow(img)
plt.title('Original Image')
plt.subplot(122), imshow(img_resized)
plt.title('Resized Image')


#print shape
print('Image shape:', img.shape)
print('resized shape:', img_resized.shape)

plt.show()
```

#4. Rescaling (Upscale/Downscale) Images using skimage
`Rescaling images is another common computer vision technique. This implies scaling the images "by a particular factor". For example  reducing the size of each image by half (downscale) or increasing the size of images by a factor of 2 (upscale).`

`when we do scaling(upscaling or downscaling) ,then a new image with a higher or lower number of pixels must be generated.`


`In downscaling we have low magnitude in digital image.`

```
from skimage.transform import rescale
img = imread('tiger.jpeg')
img_rescaled = rescale(img, scale=(0.5, 0.5))


plt.subplot(121), imshow(img)
plt.title('Original Image')

plt.subplot(122), imshow(img_rescaled)
plt.title('Rescaled Image')

#print shape
print('Image shape:', img.shape)
print('img_rescaled shape:', img_rescaled.shape)

plt.show()
```

```
from skimage.transform import rescale
img = imread('tiger.jpeg')
img_rescaled = rescale(img, scale=(0.1, 0.1))


plt.subplot(121), imshow(img)
plt.title('Original Image')

plt.subplot(122), imshow(img_rescaled)
plt.title('Rescaled Image')

#print shape
print('Image shape:', img.shape)
print('img_rescaled shape:', img_rescaled.shape)

plt.show()
```

#6. Flip Images Horizontally and Vertically
`We can flip an image both horizontally and vertically. This creates a mirror image along the horizontal/vertical axis. We can use this technique for both image preprocessing and image augmentation.`

`NumPy provides functions flipud and fliplr for flipping the images across the horizontal and vertical axis respectively.`

```
#Vertically flip (creates mirror along the y-axis)
from numpy import fliplr, flipud
dog = imread('dog0.jpg')
cat = imread('cat.jpg')

dog_flip = fliplr(dog)
cat_flip = fliplr(cat)

plt.subplot(141), imshow(dog)
plt.subplot(142), imshow(dog_flip)
plt.subplot(143), imshow(cat)
plt.subplot(144), imshow(cat_flip)
plt.show()
```

```
#Horizontallyflip   (create mirror along the x axis)
from numpy import fliplr, flipud
dog = imread('dog0.jpg')
cat = imread('cat.jpg')

dog_flip = flipud(dog)
cat_flip = flipud(cat)

plt.subplot(141), imshow(dog)
plt.subplot(142), imshow(dog_flip)
plt.subplot(143), imshow(cat)
plt.subplot(144), imshow(cat_flip)
plt.show()
```

#8. Altering Image Brightness using skimage

`Images with different brightness can be used to make our computer vision model robust to changes in lighting conditions.`

`This is important for systems that work in outdoor lighting, for instance, CCTV cameras on traffic signals.`

`The brightness of images can be changed using the adjust_gamma function in skimage, which uses a method called gamma correlation. For any given image, the pixel values are first normalized between 0  1 and then multiplied by a specified gamma value. The resulting pixel values are scaled back to the range 0-255.`

`For gamma greater than 1, the output image will be darker than the input image. While for gamma less than 1, the output image will be brighter than the input image.`

```
from skimage import exposure

#adjusting brightness
image = imread('tiger.jpeg')
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5,gain=1)

# plotting images
plt.subplot(131), imshow(image)
plt.title('Original Image')

plt.subplot(132),imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133),imshow(image_dark)
plt.title('Dark Image')

plt.show()
```

#9. Using Filters in skimage
`We can use Filters to modify or enhance an images features.`

`We can use filters for various purposes, such as smoothing and sharpening the image, removing noise, highlighting features and edges in the image, etc.`

`When we apply a filter on an image, every pixel value is replaced by a new value generated using surrounding pixel values. The simplest filter is the median filter, where the pixel values are replaced with the median of neighboring pixels.`

```
from skimage.filters import median

image = imread('tiger.jpeg', as_gray=True)
image_median = median(image)

# plotting images
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_median)
plt.title('Smooth Image')

plt.show()
```

`Let's write some code in python to create a feature matrix. where all values are initialized to 0. This matrix will store the mean pixel values for the three channels:`

```
image = imread('dog0.jpg')
feature_matrix = np.zeros((700,700)) 
feature_matrix.shape , feature_matrix[:5].shape,feature_matrix[:5]
```

```
for i in range(0,image.shape[0]):
    for j in range(0,image.shape[1]):
        feature_matrix[i][j] = ((int(image[i,j,0]) + int(image[i,j,1]) + int(image[i,j,2]))/3)  # Calculate the mean of pixels
        
```

`The new matrix will have the same height and width but only 1 channel. Now we can follow the same steps that we did in the previous section. We append the pixel values one after the other to get a 1D array:`

```
feature_matrix.shape , feature_matrix[:5]
```
