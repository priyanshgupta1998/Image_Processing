```

  
def main(): 
    try: 
        #Relative Path ,, Open a particular image from a path:        
        img = Image.open(image)  
        
        #Getting histogram of image 
        print img.histogram() 
        
        
        #transposing image  
        transposed_img = img.transpose(Image.FLIP_LEFT_RIGHT) 
        
        #Retrieve size of image
        width, height = img.size
        img = img.resize((width/2, height/2))
        
        #Create a Thumbnail In-place modification 
        img.thumbnail((200, 200))  
          
        #Angle given 
        img = img.rotate(180)  
          
        #Crop the image  
        area = (0, 0, width/2, height/2) 
        img = img.crop(area)
         #Saved in the same relative location 
        img.save("rotated_picture.jpg") 
        
        #Pate another image with current image , Image which we want to paste 
        img2 = Image.open("picture2.jpg")  
        img.paste(img2, (50, 50)) 
        
        
        
    except IOError: 
        pass
  
if __name__ == "__main__": 
    main()
```
