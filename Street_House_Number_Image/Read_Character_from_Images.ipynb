```
from sklearn.utils import shuffle
import numpy as np
def arr(fold):
  Images = []
  for image_file in fold:
#     print(fold.index(image_file) , ' ---> ' , image_file)
    if('.png' in image_file):
      image = cv2.imread('/content/train/' + image_file) #Reading the image (OpenCV)
      image = cv2.resize(image,(150,150)) #Resize the image, Some images are different sizes. (Resizing is very Important)
      Images.append(image)
  return shuffle(Images,random_state=817328462) #Shuffle the dataset you just prepared.

Images = arr(train_folder)
print(len(Images))  # two image were not in proper format i just removed from the traning folder 
```
