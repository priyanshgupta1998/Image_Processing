```
import pydicom
import matplotlib.pyplot as plt
img = '/content/IMAGES/' + dicom[0]
plt.imshow(pydicom.dcmread(img).pixel_array)
plt.show()
```

```
import pydicom
import matplotlib.pyplot as plt
img = '/content/IMAGES/' + dicom[0]
plt.imshow(pydicom.dcmread(img).pixel_array, cmap=plt.cm.bone)  # we can pass two extra argument here (vmin=-200 , vmax = 200)
plt.show()
```
